### Importing Dependencies

```python
from random import choice
from random import randint as rnd
from random import shuffle
import time 
```

### Setting the problem parameters

```python
ALPHABET = [chr(i) for i in range(97,123)]+[" "]
TRUE_TEXT = "hello and welcome to the numpy tutorial"
FIXED_SCORE = [ord(i) for i in TRUE_TEXT]
FIXED_SCORE = sum(FIXED_SCORE)
N = len(TRUE_TEXT)
MAX_POPULATION = 200
MUTATION_RATE = 0.7
EPOCH = 10000
E = EPOCH
```

### Initialize Population function

```python
def init_population(n,m):
    population_list = []
    for j in range(m):
        member = [choice(ALPHABET) for i in range(n)]+[None]
        population_list.append(member)
    return population_list
```

### Cross over function

```python
def cross_over(population_list,n,m):
    half = n//2
    for i in range(m):
        child = population_list[i][:n].copy()
        child+=[None]
        population_list.append(child)
    return population_list
```

### Mutation function

```python
def mutation(population_list,n,m,r):
    choosen_ones = [i for i in range(m,m*2)]
    shuffle(choosen_ones)
    choosen_ones = choosen_ones[:int(m*r)]
    for child in choosen_ones:
        cell1 = rnd(0,n-1)
#         cell2 = rnd(0,n-1)
        population_list[child][cell1]=choice(ALPHABET)
#         population_list[child][cell2]=choice(ALPHABET)
    return population_list
```

### Fitness function

```python
def fitness(population_list,n,f):
    for item in population_list:
        collison = 0
        for i in range(len(item)-1):
            if item[i] == TRUE_TEXT[i]:
                pass
            else:
                collison+=1
        item[-1] = collison
    return population_list
```

### Main

```python
print(N)
print(TRUE_TEXT)
current_population = init_population(N,MAX_POPULATION)
current_population = fitness(current_population,N,FIXED_SCORE)
current_population = sorted(current_population, key=lambda x:x[N])
if current_population[0][N]==0:
    print(f"The given text has been found in the initial population!{current_population[0]}")
else:
    while EPOCH:
        current_population = cross_over(current_population,N,MAX_POPULATION)
        for i in range(3):
            current_population = mutation(current_population,N,MAX_POPULATION,MUTATION_RATE)
        current_population = fitness(current_population,N,FIXED_SCORE)
        current_population = sorted(current_population, key=lambda x:x[N])
        current_population = current_population[:MAX_POPULATION]
        EPOCH-=1
        if current_population[0][N]==0:
            print("#",E-EPOCH,"Solution Found:",current_population[0])
            print("".join(current_population[0][:N]))
            break
        else:
            print("#",E-EPOCH,"Best Solution so far:",current_population[0])
    
    else:
        print("Sorry We Couldnt find a solution for you.")
```
